```python
import numpy as np
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
from PIL import Image
from keras import utils
sys.path.append('../rxrx1-utils')
import rxrx.io as rio
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

```python
from skimage.transform import match_histograms
```

```python
def parse_all_train(train_csv_path='', 
                    target_dir='', 
                    use_control_eq=False, 
                    RGB=False, 
                    num_samples=-1, 
                    n_sirna=-1,
                    **kwargs):
    '''
    Description: Parses and saves out training data.
    
    Args:
        train_csv_path = .csv path for the training file
        RGB = whether or not to convert the 6-channel image to RGB
        num_samples = number of samples to consider (set to -1 to use all available samples)
        n_sirna = number of siRNA to consider (set to -1 for all classes)
    
    Outputs:
        x_id: IDs for x_data arrays 
        y_id: IDs for y_data arrays
    
    '''
    
    #read in train CSV file
    train_df = pd.read_csv(train_csv_path)
    control_df = pd.read_csv(kwargs['control_path'])
    control_df = control_df.loc[control_df['well_type'] == 'positive_control']
    control_df = control_df.loc[control_df['sirna'] == 1109]
    
    if n_sirna != -1:
        train_df = train_df.loc[train_df['sirna'] < n_sirna]
    
    if num_samples == -1:
        num_samples = len(train_df)
        print(f"Reading all samples from dataset: {num_samples}")
    
    #get targets to write out
    num_classes = len(train_df.sirna.unique())
    y_train = train_df['sirna'].values
    y_train = utils.to_categorical(y_train, num_classes=num_classes)
    print("y_train shape: ", y_train.shape)
    
    #get df values
    dataset = 'train'
    experiments = train_df['experiment'].values
    plates = train_df['plate'].values
    wells = train_df['well'].values
    sites = np.array([1,2])
    channels=np.array([1,2,3,4,5,6])
    
    #get control df values
#     experiments_control = control_df['experiment'].values
#     plates_control = control_df['plate'].values
    wells_control = control_df['well'].values
    
    x_id = []  #list to compile x IDs
    y_id = []  #list to compile y IDs
    
    #instantiate StandardScaler obj
    scaler = StandardScaler()
    
    for site in sites:
        
        #initialize j counter (for control wells)
        j=0
        
        for i in tqdm_notebook(range(num_samples)):
            
            site_img = np.empty((512, 512, 6), dtype=np.float32)
            
            if use_control_eq:
                control_img = np.empty((512, 512, 6), dtype=np.float32)
                
            for channel in channels:
                path = f'../../data/{dataset}/{experiments[i]}/Plate{plates[i]}/{wells[i]}_s{site}_w{channel}.png'
                site_img[:, :, channel-1] = Image.open(path)
            
                if use_control_eq:
                    cntrl_path = f'../../data/{dataset}/{experiments[i]}/Plate{plates[i]}/{wells_control[j]}_s{site}_w{channel}.png'                    
                    control_img[:, :, channel-1] = Image.open(cntrl_path)
                    site_img[:, :, channel-1] = match_histograms(site_img[:, :, channel-1], control_img[:, :, channel-1], multichannel=False)
                
            if RGB:
                site_img = np.asarray(rio.convert_tensor_to_rgb(site_img, vmax=255), dtype=np.float16)
            
            #normalize
            site_img /= 255
#             site_img = scaler.fit_transform(site_img)
            
            #write out the image
            np.save(f'{target_dir}x_{experiments[i]}_{plates[i]}_{wells[i]}_s{site}', site_img, allow_pickle=True)
            np.save(f'{target_dir}y_{experiments[i]}_{plates[i]}_{wells[i]}_s{site}', y_train[i], allow_pickle=True)
            
            x_id.append(f'x_{experiments[i]}_{plates[i]}_{wells[i]}_s{site}')
            y_id.append(f'y_{experiments[i]}_{plates[i]}_{wells[i]}_s{site}')
            
            #increment the wells controls index
            if i != num_samples-1:
                if plates[i] != plates[i+1]: j += 1 
    
    return x_id, y_id
```

```python
x_id, y_id = parse_all_train(train_csv_path='../../data/train.csv',
                             target_dir='../../data/train_parsed_first_50_hist_matched/',
                             use_control_eq=True,
                             RGB=False, 
                             num_samples=-1,
                             n_sirna=50,
                             control_path='../../data/train_controls.csv')

np.save('../../data/x_id_first_50_hist_matched', x_id, allow_pickle=True)
np.save('../../data/y_id_first_50_hist_matched', y_id, allow_pickle=True)
```
